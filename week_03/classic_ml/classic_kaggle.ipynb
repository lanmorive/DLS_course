```python
from sklearn.compose import ColumnTransformer,make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder,StandardScaler,Normalizer,RobustScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,StratifiedKFold,cross_val_score,GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False), cat_cols),
        ('num', MinMaxScaler(), num_cols)
    ]
)

pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])



cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

param_grid = [
    {
        'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [LogisticRegression(solver='liblinear',max_iter=1000)],
        'classifier__penalty': ['l1', 'l2'],
        'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
    },
    {   'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [LogisticRegression(solver='saga', max_iter=1000)],
        'classifier__penalty': ['elasticnet'],
        'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100],
        'classifier__l1_ratio': [0, 0.25, 0.5, 0.75, 1]
    },
    {   'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [LogisticRegression(solver='saga', max_iter=1000)],
        'classifier__penalty': ['l1', 'l2', 'none'],
        'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
    },
    {
        'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [LogisticRegression(solver='lbfgs',max_iter=1000)],
        'classifier__penalty': ['l2', 'none'],
        'classifier__C': [0.01, 0.1, 1, 10, 100]
    },
    {
        'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [LogisticRegression(solver='newton-cg',max_iter=1000)],
        'classifier__penalty': ['l2', 'none'],
        'classifier__C': [0.01, 0.1, 1, 10, 100]
    },
    {
        'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [GaussianNB()]
    },
    {
        'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [KNeighborsClassifier()],
        'classifier__n_neighbors': [3, 5, 7, 9],
        'classifier__weights': ['uniform', 'distance'],
        'classifier__p': [1, 2]
    },
    {
        'preprocessor__num': [MinMaxScaler(), RobustScaler(), StandardScaler(), Normalizer()],
        'classifier': [RandomForestClassifier()],
        'classifier__n_estimators': [50, 100, 200,300],
        'classifier__max_depth': [3,4,5,6,7,8]
    }
]




grid = GridSearchCV(
    pipe,
    param_grid,
    scoring='roc_auc',
    cv=cv,
    n_jobs=-1,
    verbose=1
)
```

```python
X = train_data[feature_cols]
y = train_data[target_col] 
```

```python
grid.fit(X, y)
```

```python
grid_result = pd.DataFrame(grid.cv_results_).sort_values(["rank_test_score",'std_test_score']).T
grid_result.head(30)
```

```python
grid.best_params_
```

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметров количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию.

```python
X_train, X_val, y_train, y_val = train_test_split(train_data[feature_cols], train_data['Churn'], test_size=0.2, random_state=42, stratify=y)
```

```python
from catboost import CatBoostClassifier

cat_model = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.1,
    depth=6,
    eval_metric='AUC',
    random_seed=42,
    verbose=20
)

cat_model.fit(
    X_train, y_train,
    cat_features=cat_cols,
    eval_set=(X_val, y_val),

)
```

```python
pd.read_csv('submission (2).csv').assign(Churn=cat_model.predict_proba(test_data)[:, 1]).to_csv('submission.csv', index=False)
```
